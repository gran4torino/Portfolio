Загружаем данные в датафрейм с названием города и его населением из таблицы City

```python
df_city = pd.read_sql(
"""
SELECT
 name, population
FROM
 city
""",
"postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db"
)
df_city
```

Cоединяем данные из выгруженного CSV файла (df_csv) и таблицы с населением города df_city

```python
df_merged = df_city.merge(
    df_csv,
    left_on='name', 
    right_on='Сity', 
    how='right')
df_merged
```

```python
# группируем по странам и рассчитываем среднюю численность населения в городах, в которых я отдыхал

df_new = df_merged.groupby('Страна отдыха?').mean()
df_new['population']
```
