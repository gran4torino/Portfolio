```python
import pandas as pd
import glob

#import matplotlib.pyplot as plt
#import seaborn as sns

#задаем путь для папки, где у нас выложены дневные отчеты по заболеванию Covid-19, читаем все файлы из этой папки
path = r'/content/drive/MyDrive/Netology/Диплом/Данные/Данные/Kaggle Covid/csse_covid_19_data/csse_covid_19_daily_reports' 

all_files = glob.glob(path + "/*.csv")

df = pd.concat((pd.read_csv(f) for f in all_files))
df.head()
```

```python
df.info()
```

```python
#Данных по столбцам Recovered и Active (501) меньшt, чем в столбцах 3-8 (714). Найдем дату публикации последних значений по столбцу Recovered. Это 5.08.2021 г. Отсутствие данных
# после этой даты объявлено автором в описании полей к данному набору данных.
df.dropna(subset=['Recovered']).sort_values('Last_Update').tail(10)
```

```python
#переведем в формат datetime столбец Last_Update. Нормализуем дату (изменим все время на 00:00:00)
df['Last_Update'] = pd.to_datetime(df['Last_Update']).dt.normalize()
#отфильтруем датафрейм и оставим только данные по Польше
df = df.loc[df['Country_Region'] == 'Poland']
# удалим лишние столбцы, сохраним датафрейм под новым названием df_covid
df_covid = df.drop(columns = ['FIPS' ,'Admin2', 'Province_State', 'Country_Region', 'Lat',  'Long_','Active' , 'Case_Fatality_Ratio', 'Province/State', 'Country/Region', 'Last Update', 'Latitude', 'Longitude','Incident_Rate',  'Incidence_Rate', 'Case-Fatality_Ratio', 'Combined_Key' ])
df_covid.info()
```

```python
# Обогатим датафрейм данными с количеством привитых в Польше. Для этого загрузим данные в датафрейм (источник- Министерство здоровья Польши)
df_vaccinations = pd.read_csv('/content/drive/MyDrive/Netology/Диплом/Данные/Данные/Данные по прививкам от Covid-19 Польша/20211128075012_rap_rcb_daily_szczepienia.csv',
                              sep = ';')
df_vaccinations.head(20)
```

```python
#последние данные по прививкам - 27 ноября 2021 года
df_vaccinations.tail(20)
```

```python
df_vaccinations.info()
```

```python
#переведем в формат datetime столбец data
df_vaccinations['data'] = pd.to_datetime(df_vaccinations['data'])
df_vaccinations.info()
```

```python
# переименуем столбцы на английский язык
df_vaccinations =  df_vaccinations.rename(columns={"data":"Last_Update", "liczba_szczepien_dziennie": "Total_vaccination_quantity", 
                                                   "dawka_2_dziennie":"Second_vaccination_quantity", "dawka_1_dziennie" : "First_vaccination_quantity",
                                                   "dawka_3_dziennie" : "Third_vaccination_quantity",
                                                   "dawka_przypomianjaca_dziennie" :  "Reminding_vaccination_quantity"
                                                   })
df_vaccinations.info()
```

```python
# Оставим только общее количество людей, сделавшее прививки (Total_vaccination_quantity)
df_vaccinations = df_vaccinations[[
 'Last_Update',
 'Total_vaccination_quantity'
]]
df_vaccinations.info()
```

```python
#Объединим датафреймы с ковидом df_covid и df_vaccinations по столбцу с датой
df_covid = df_covid.merge(df_vaccinations, how='outer', on='Last_Update')
df_covid.head(10)
```

```python
#Проверим на валидность объединения- все данные объединились корректно, количество строк в исходном датафрейме df_covid в стольцах 0-4 и 5-14 соответствуеи исходным датафреймам 
df_covid.info()
```

```python
# загрузим данные по ценам недвижимости на первичном рынке в датафрейм
df_realt = pd.read_csv('/content/drive/MyDrive/Netology/Диплом/Данные/Данные/Цены на недвижимость Польша/Realt prices on Poland for analysis.csv', 
                       sep=';')
df_realt.head()
```

```python
df_realt.info()
```

```python
#переименуем на английский язык некоторые столбцы со средней ценой и удалим один столбец со средней ценой по 9 городам 
df_realt =  df_realt.rename(columns={"7 miast":"price_average_7_first_biggest_cities", "10 miast": "price_average_10_last_biggest_cities", 
                                     "6 cities without Warszawa": "price_average_6_biggest_cities_w_o_Warszawa"
                                     })
df_realt = df_realt.drop(columns = ['9 cities'])
df_realt.info()
```

```python
#поменяем формат столбца год-квартал на период квартальных данных period[Q-DEC]
df_realt['Quarter, year']=df_realt['Quarter, year'].astype('period[Q]')
df_realt.info()
```

Для корреляционного анализа необходимо объединить датафрейм с данными по covid и датафрейм с ценами на недвижимость.
Так как в датафрейме с данными цен на недвижимость размерность дат - кварталы, а в датафрейме с переменными Covid - дни, добавим столбец с годом  и кварталом в датафрейм с Covid.

```python
# добавим столбцы с годами и кварталами в датафрейм с Covid
df_covid['Quarter, year'] = pd.to_datetime(df_covid['Last_Update'].values, format='%Y-%m').astype('period[Q]')
df_covid.head()
```

```python
#перенесем столбец Quarter, year к столбцу Last_Update
df_covid = df_covid[[
 'Last_Update',
 'Quarter, year',
 'Confirmed',
 'Deaths',
 'Recovered',
 'Total_vaccination_quantity',
 ]]
df_covid.info()
```

```python
#Объединим датафреймы с ковидом и ценами на недвижимость по столбцу с квартальными данными Quarter, year. Создадим новый датафрейм df_full.
df_full = df_covid.merge(df_realt, how='outer', on='Quarter, year')
df_full.head(10)
```

```python
#Посмортим, какие самые ранние данные у нас есть по ценам на недвижимость - ответ с третьего квартала 2006 года.
df_full.sort_values('Quarter, year').head(10)
```

```python
#Посмотрим с какой даты есть данные по covid- ответ с 22.03.2020 - значит данные есть с начала эпидемии за пределами Китая
df_full.sort_values('Last_Update').head()
```

```python

#Оставим только данные после  22.03.2020 - с этой даты есть данные по ковиду и ценам на недвижимость
#и удалим данные позднее, чем 31.12.2021 - последние данные по ценам на недвижимость есть за 4-й квартал 2021 года.
df_full = df_full[(df_full['Last_Update'] >= '2020-03-22') & (df_full['Last_Update'] < '2022-01-01') ]
df_full.sort_values('Last_Update').tail(10)
df_full.info()
```
